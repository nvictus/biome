# Examples of microbiome visualizations

## Initialize and load the data

```python
import pandas
%pylab inline
%load_ext rmagic

df = pandas.read_csv('data/relman.txt', header=0, delimiter='\t')
```

## Snapshot of one subject

```python
grp = df['D1'].groupby(df.Phylum)
#grp.ngroups
#report = grp.describe()
rel_abd = grp.sum() / np.sum(grp.sum())
print rel_abd
print sum(rel_abd)
```

```python
labels = [
    '4C0d-2', 
    'Actinobacteria',
    'Bacteroidetes',
    'TM7',
    'Cyanobacteria',
    'Firmicutes',
    'Fusobacteria',
    'Lentisphaer*',
    'Proteobacteria',
    'Synergist*',
    'Verrucomicrobia'
]
```

```python
explode = np.zeros(11)
explode[5] = 0.2
plt.pie(rel_abd, labels=labels, startangle=0, explode=explode, autopct='%1.1f%%')
plt.title('matlplotlib defaults are ugly!!!');
```

## Compare subject snapshots

```python
data = []
for subject in ['D1', 'E1', 'F1']:
    grp = df[subject].groupby(df['Phylum'])
    data.append(grp.sum()/sum(grp.sum()))
```

```python
%%R -i labels,data

library(ggplot2)

df <- data.frame(labels=labels, D=data[1,], E=data[2,], F=data[3,])
x <- c(rep('D', 11), rep('E', 11), rep('F', 11))
y <- c(df$D, df$E, df$F)
phylum <- c(df$labels, df$labels, df$labels)

p <- ggplot(df, aes(x=x, y=y, fill=phylum)) + 
    geom_bar(stat="identity", width=0.8) +
    xlab("Subject") +
    ylab("Abundance") + 
    ggtitle("Comparing 3 subjects")

print(p)
```

```python
grp = df.groupby(df.Phylum)
df2 = grp.sum()/sum(grp.sum())
abundances = df2[[i for i in df2 if i.startswith('D')]]
num_times = abundances.shape[1]
x = reduce(lambda x,y: x+y, [[i]*11 for i in range(1,56+1)], [])
y = abundances.values.T.flatten()
labels2 = reduce(lambda x,y: x+y, [labels for i in range(56)], [])
```

## Track subjects over time

```python
%%R -i x,y,labels2

library(ggplot2)

df = data.frame(x=x, y=y, phylum=labels2)
p <- ggplot(df, aes(x=x, y=y, fill=phylum)) + 
    geom_bar(stat="identity", width=0.8) +
    ggtitle('one subject over time') +
    xlab('time') +
    ylab('relative abundance')
print(p)
```

### Firm-to-Bact ratio

```python
grp = df.groupby(df.Phylum)
bact = grp.get_group('Bacteroidetes')
firm = grp.get_group('Firmicutes')
F2B = firm.sum()/bact.sum()

plt.plot(F2B[ [i for i in F2B.index if i.startswith('D')] ])
plt.plot(F2B[ [i for i in F2B.index if i.startswith('E')] ])
plt.plot(F2B[ [i for i in F2B.index if i.startswith('F')] ])
plt.xlabel('time')
plt.ylabel('Firm/Bac')
```
